The functions below will download the dataset into your browser:



```python
# Define the URL of the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
## Write your code here
df.info()
df.describe(include='all')
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
df.isnull().sum()
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
import seaborn as sns
import matplotlib.pyplot as plt

# Set the figure size for better readability
plt.figure(figsize=(12, 6))

# Create the heatmap
sns.heatmap(df.isnull(), 
            cbar=False, 
            cmap='viridis', 
            yticklabels=False)

# Add a title
plt.title("Heatmap of Missing Values", fontsize=16)

# Show the plot
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
missing_RemoteWork = df['RemoteWork'].isnull().sum()
print(f"Number of missing rows in 'RemoteWork': {missing_RemoteWork}")
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
majority_value = df['RemoteWork'].mode()[0]
print(f"The most frequent value in 'Employment' is: {majority_value}")
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
# Step 1: Find the most frequent value (mode)
most_frequent = df['RemoteWork'].mode()[0]

# Step 2: Fill missing values with that mode
df['RemoteWork'].fillna(most_frequent, inplace=True)

# Optional: Confirm that missing values are handled
print(f"Missing values after imputation: {df['RemoteWork'].isnull().sum()}")
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here
# Set figure size for clarity
# Set the figure size for better readability
plt.figure(figsize=(12, 6))

# Create the heatmap
sns.heatmap(df.isnull(), 
            cbar=False, 
            cmap='viridis', 
            yticklabels=False)

# Add a title
plt.title("Heatmap of Missing Values", fontsize=16)

# Show the plot
plt.show()
```
